## Calculating Averages and Entropy

```python
def unit_vector(vector):
    """ Returns the unit vector of the vector.  """
    return vector / np.linalg.norm(vector)
```

```python
def angle_between(v1, v2):
    """ Returns the angle in radians between vectors 'v1' and 'v2'::

            >>> angle_between((1, 0, 0), (0, 1, 0))
            1.5707963267948966
            >>> angle_between((1, 0, 0), (1, 0, 0))
            0.0
            >>> angle_between((1, 0, 0), (-1, 0, 0))
            3.141592653589793
    """
    if v1 is None or v2 is None:
        return (None, None)
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    cos = np.clip(np.dot(v1_u, v2_u), -1.0, 1.0)
    if np.isnan(np.sum(cos)):
        return None 
    return np.arccos(cos)*57.2958 , cos
```

```python
def var_avg_style(embeddings):
    variances, lengths, averages = {}, {}, {}
    for item in embeddings:
        differences_cos = []
        for painting_1 in embeddings[item]:
            for painting_2 in embeddings[item]:
                differences_cos.append(angle_between(painting_1, painting_2)[0])
        mean_cos = np.mean(differences_cos)

    variances[item] = mean_cos
    averages[item] = np.mean(embeddings[item], axis=0)
    return variances, averages
```

## Search and Clustering Based on Style




```python
import operator
```

```python
def find_similar(name, dictionary, n=5, other_dict=None, reverse=False):
    angles = {}
    if name in dictionary:
        vec = dictionary[name]
    else:
        vec = other_dict[name]
    for entity in dictionary:
        if entity != name:
            angles[entity] = round(angle_between(dictionary[entity], vec)[0], 3)
    
    sorted_n = sorted(angles.items(), key=operator.itemgetter(1))
    if not reverse:
        return(sorted_n[0:n])
    if reverse:
        sorted_n.reverse()
        return(sorted_n[0:n])
```

## Visualising Vectors with TSNE

```python
from sklearn.manifold import TSNE
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
def tsne_plot_similar_words(title, labels, embedding_clusters,
                            entity_clusters, a, filename=None, scatter_size=1):
    plt.figure(figsize=(16, 9))
    colors = cm.rainbow(np.linspace(0, 1, len(labels)))
    for label, embeddings, words, color in zip(labels, embedding_clusters, entity_clusters, colors):
        x = embeddings[:, 0]
        y = embeddings[:, 1]
        plt.scatter(x, y, color=color, alpha=a, label=label, s=scatter_size)
        for i, word in enumerate(words):
            plt.annotate(word,
                       alpha=a,
                       xy=(x[i], y[i]),
                       xytext=(5, 2),
                       textcoords='offset points',
                       ha='right',
                       va='bottom',
                       size=12,
                       color='black')
        plt.legend(bbox_to_anchor=(1.15, 1.15), prop={'size': 15})
        plt.xlabel("tSNE 0th dimensions", fontsize=15)
        plt.ylabel("tSNE 1st dimensions", fontsize=15)

        plt.title(title, fontsize=16)
    if filename:
        plt.savefig(filename, bbox_inches='tight')
    plt.show()
```

```python
def get_cluster_embeddings(entity_keys, vectors, diff_vectors=None):
    embedding_clusters = []
    entity_clusters = []
    # find closest words and add them to cluster
    for entity in entity_keys:
        embeddings = []
        entities = []
        try:
          embeddings.append(vectors[entity])
          entities.append(entity)
        except KeyError:
          embeddings.append(diff_vectors[entity])
          entities.append(entity)          
        for similar_entity, _ in find_similar(entity, vectors, other_dict=diff_vectors, n=5):
            entities.append(similar_entity)
            embeddings.append(vectors[similar_entity])
        
        embedding_clusters.append(embeddings)
        entity_clusters.append(entities)
    # get embeddings for the words in clusers
    embedding_clusters = np.array(embedding_clusters)
    # print(embedding_clusters)
    n, m, k = embedding_clusters.shape
    tsne_model_en_2d = TSNE(perplexity=10, n_components=2, init='pca', n_iter=3500, random_state=32)
    # embeddings_en_2d = tsne_model_en_2d.fit_transform(embedding_clusters.reshape(n * m, k))
    embeddings_en_2d = np.array(tsne_model_en_2d.fit_transform(embedding_clusters.reshape(n * m, k))).reshape(n, m, 2)

    return embeddings_en_2d, entity_clusters
```

```python
artist_entities = [('Pablo Picasso', 1920), ('Vincent van Gogh', 1886)]
```

```python
embeddings_en_2d_genre, entity_clusters_genre = get_cluster_embeddings(artist_entities, artist_avg_style)
```

```python
tsne_plot_similar_words('Similar Artist Entities', artist_entities, embeddings_en_2d_genre, entity_clusters_genre, 1.0, scatter_size=25, filename="artist.pdf")
```

```python
def aggregate_embeddings(embeddings):
    agg_embedding = {}
    for embedding in embeddings:
        emb, year = embedding
        if emb not in agg_embedding:
            agg_embedding[emb] = []
        if emb in agg_embedding:
            for embed in embeddings[embedding]:
                if not np.isnan(embed.any()):
                    agg_embedding[emb].append(embed)
    return agg_embedding
```

```python
agg_artist_style_embedding = aggregate_embeddings(artist_style_embeddings)
```

#### PCA analysis


```python
acc_matrix = []
acc_label = []
counter = 0

for label, embeddings in agg_artist_style_embedding.items():
    for embed in embeddings: 
        acc_matrix.append(embed)
        acc_label.append(label)
matrix = np.stack(acc_matrix)
labels = np.array(acc_label)
```

```python
import sklearn.decomposition  as dec
import matplotlib.pyplot as plt
import seaborn as sns
model = dec.PCA(4)
matrix_red = model.fit_transform(matrix)
```

```python
from collections import Counter
fig, axs = plt.subplots(1, 1, figsize=(8, 8))
most_common = Counter(labels).most_common(10)

for k, _ in most_common:
    x = matrix_red[labels == k, 0].mean()
    y = matrix_red[labels == k, 1].mean()
    axs.scatter(x, y, label=k)
    axs.text(x + 0.002, y + 0.002, s=k, ha="left")

plt.xlabel("Dimension 0 of PCA")
plt.ylabel("Dimension 1 of PCA")
plt.title("PCA Analysis of the 10 most represented Artists")
plt.show()
```
