```python
import json

def parse_json_to_dict(filename):
    with open (filename, 'r') as read_file:
        data = json.load(read_file)
    return data

def save_dict_to_json(mydict, filename):
    json_dict = json.dumps(mydict)
    f = open(filename, 'w')
    f.write(json_dict)
    f.close()
```

```python
artist_genres = parse_json_to_dict("artist_genres.json")
```

```python
artist_styles = parse_json_to_dict("artist_styles.json")
```
